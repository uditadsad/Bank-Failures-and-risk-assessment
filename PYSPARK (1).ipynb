```python
from pyspark.sql.functions import corr

def check_multicollinearity(df, target_columns, threshold=0.75):
    cols_to_drop = []
    numeric_cols = [col for col, dtype in df.dtypes if dtype in ['int', 'bigint', 'float', 'double']]
    cols_to_check = [col for col in numeric_cols if col not in target_columns]

    for i in range(len(cols_to_check)):
        for j in range(i+1, len(cols_to_check)):
            col1 = cols_to_check[i]
            col2 = cols_to_check[j]
            correlation = df.select(corr(col1, col2)).collect()[0][0]
            # if abs(correlation) > threshold:
            #     cols_to_drop.append(col2)
            #     print(f"Multicollinearity detected between {col1} and {col2}. Correlation: {correlation}")
            if correlation is not None and abs(correlation) > threshold:
                cols_to_drop.append(col2)
                print(f"Multicollinearity detected between {col1} and {col2}. Correlation: {correlation}")

    return cols_to_drop

# List of target columns to exclude from checking
target_columns = ['FailWithin90Days', 'FailWithin180Days', 'FailWithin365Days', 'FailWithin730Days']

# Call the function
columns_to_drop = check_multicollinearity(merged_df, target_columns)
print("Columns to drop due to multicollinearity:", columns_to_drop)
```

```python
from pyspark.sql.functions import when, mean

def fill_missing_values(df):
    for column in df.columns:
        if df.schema[column].dataType == 'string':
            # For string columns, fill missing values with mode
            mode_value = df.groupBy(column).count().orderBy('count', ascending=False).first()[0]
            df = df.withColumn(column, when(df[column].isNull(), mode_value).otherwise(df[column]))
        else:
            # For numerical columns, fill missing values with mean
            mean_value = df.select(mean(col(column))).collect()[0][0]
            df = df.withColumn(column, when(df[column].isNull(), mean_value).otherwise(df[column]))
    return df

# Fill missing values in merged_df
merged_df = fill_missing_values(merged_df)
```

# Decission tree model

```python
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

# Define the DecisionTreeClassifier model
decision_tree = DecisionTreeClassifier(featuresCol='features', labelCol='Failwithin180days')

# Fit the model
decision_tree_model = decision_tree.fit(train_data)

# Make predictions on validation data
validation_predictions = decision_tree_model.transform(validation_data)

# Make predictions on test data
test_predictions = decision_tree_model.transform(test_data)

# Evaluate the model on validation data
binary_evaluator = BinaryClassificationEvaluator(labelCol='FailWithin180days', metricName='areaUnderROC')
multiclass_evaluator = MulticlassClassificationEvaluator(labelCol='FailWithin180days', metricName='f1')

auc_validation = binary_evaluator.evaluate(validation_predictions)
f1_scores_validation = multiclass_evaluator.evaluate(validation_predictions)

print(f'AUC (Validation): {auc_validation}')
print(f'F1 Score (Validation): {f1_scores_validation}')

# Evaluate the model on test data
auc_test = binary_evaluator.evaluate(test_predictions)
f1_scores_test = multiclass_evaluator.evaluate(test_predictions)

print(f'AUC (Test): {auc_test}')
print(f'F1 Score (Test): {f1_scores_test}')
```
